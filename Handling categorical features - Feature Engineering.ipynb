# Handle categorical features

There are two types of categorical features:

Nominal category - Feature where we do not have to worry about the arrangement of categories. Example- Gender, State, Country. 

Oridnal category - We have to worry about ranks. Here, categories can arranged based on rank. Example - We want to preduct the salary and education column is present in data. Here we would see what degree person has and based on that we would give rank. PHD would have highest rank, masters would have lower than that etc... So, based on this rank, we would predict the salary.

There are two types pf encoding techniques

1. Nominal encoding - They have 4/5 techniques.
   1. One hot encoding
   2. One hot encoding with many categorical variables
   3. Mean encoding
3. Ordinal encoding :
   1. Label encoding
   2. Target guided oridnal encoding.

# 3. Ordinal Number Encoding (Label encoding) -Ordinal Categories

Where categories can be ordered such a wau where you can assign ranks. For example, grading system - A, B, C, D, F

You rank saying A can be given value 1, B can be given 2, C-3, D-4, F-5

Experience of batsman. 

A- 10 years experience -1 rank
B- 5 years experience-2nd rank
C- 3 years exp- 3rd rank

So, here we can give ranking based expirience. So, like that, in ordinal number encoding, we can give numbers/ranks to categories.

Also, weekdays and weekends. 
MON -7
TUES -6
WED -5
THURS -4
FRI -3
SAT -2
SUN -1

SO, we can give ranks something like this.

Advantages:
1. Keeps the semantic information of the variable (human readable content)
2. straightforward

```python
# Lets take example of ordinal encoding.
import pandas as pd
import datetime


# Lets find todays date.

today_date = datetime.datetime.today()
```

```python
# Lets get 15 previous days.


datetime.timedelta() #this function would give the difference between two dates.

today_date - datetime.timedelta(3) #So here, if we pass 3 in delta function, then it would subtract 3 days from todays date and would give date 3 days back from today.
```

```python
# Lets get 15 days using above function

days = [today_date - datetime.timedelta(i) for i in range (0,15) ]

```

```python
# Lets convert this into dataframe.

data = pd.DataFrame(days)
data.columns=["Day"]
```

```python
data
```

```python
# Lets get weekday name of every day

data["Weekday"] = data["Day"].dt.day_name()
```

```python
data
```

```python
# Lets give ranks to this days now.

dict_1 = {"Monday":1, "Tuesday":2, "Wednesday":3, "Thursday":4, "Friday":5, "Saturday": 6, "Sunday":7} #This is just example, you can give any rank as per your choice.

```

```python
# Lets map this dictionary to weekday
data["Weekday_ordinal"] = data["Weekday"].map(dict_1) # here, we mapped values for everyday from dictionary.
```

```python
data #see the last column, we have converted categorical data to numerical by ordinal encoding.
```

# 4. Count or Frequency encoding - Nominal Categories

We use this technique when we have higher number of categories (high cardinality).

```python
# Lets learn this technique using below data.

train_set = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None)
```

```python
train_set
```

```python
train_set.dtypes
```

```python

columns = [column for column in train_set.columns if train_set[column].dtypes=="object" ]
# categorical_data = pd.DataFrame(np.array([train_set[column] for column in train_set.columns]))
categorical_cols = train_set[columns]
```

```python
categorical_cols.columns=["Employment", "Degree", "Status", "Designation", "Family-Job", "Race", "Sex", "Country", "Cost"]
```

```python
categorical_cols
```

```python
# Now lets check how many categories are present in every column.

for column in categorical_cols.columns:

    print(column, " has ", len(categorical_cols[column].unique()), "categories")
```

```python
# Ok, now we are going to work on country column for this type of encoding.

dict_2 = categorical_cols["Country"].value_counts().to_dict()
```

```python
dict_2
```

```python
categorical_cols["Country_frequency_encoded"] = categorical_cols["Country"].map(dict_2)
```

```python
categorical_cols
```

### Above, you can see we have assigned the frequency of specific category to that specific category to perform the encoding.

## Advantages : 

1. Easy to use
2. We are not creating any new columns for categories, so not increasing feature space.

## Disadvantage : 

1. If frequency of 2 categories are same, then it would provide the same weight which is a major problem. Model would not be able to distinguish between those 2 categories.
    

# 5. Target guided ordinal encoding - Ordinal Categories

here, we would just not take the categorical feature but the output feature as well along with it. 

1. Ordering the categories according to target.
2. Replace the categories by the joint probability of being 1 or 0. 

Here, we are mainly focused on our target/depdendent variable. Now, we will take categories and would find the mean of dependent variable for every category. See the titanic datset example, you would understand. Then, once the mean for every category is found, we would sort the categories in descending order of mean. Then, we would assign rank to every category based on the mean value it has. Whichever category has the highest mean would have the highest rank and so on..Then we are going to replace this categories with the rank we have assigned them. Here, ranking of category is mainly based on target variable thats why it is called as target guided ordinal encoding.

```python
# Lets use the titanic data to see the practical implementation of this method.

data = pd.read_csv("Titanic.csv", usecols=["Survived","Cabin"])
```

```python
data
```

```python
# Lets first replace null values in cabin column with new category.

data["Cabin"].fillna("Missing", inplace=True) 
```

```python
print(data.isnull().sum())
data
```

### Now, we are going to work on cabin column as its a categorical column and survived is a dependent variable. 

```python
# Lets first just figure out the cabin from every entry as we do not cabin number, just cabin ID is enough.

data["cabin_id"] = data["Cabin"].astype(str).str[0] # Here, we have got the cabin ID directly (first letter of cabin number)
```

```python
data
```

```python
# Lets find what percentage of people are survived in every cabin and then sort them in ascending/descending order.
print(data.groupby(["cabin_id"])["Survived"].mean().sort_values())
survival_percantage = data.groupby(["cabin_id"])["Survived"].mean().sort_values().index
survival_percantage
```

```python
# In above results we can see that T has the lowest survival rate while D has the highest survival rate. Lets give rank to this categories/cabins by their survival percantage.

for i,k in enumerate(survival_percantage,0):
    print(i,k)
```

```python
# Now lets assign ranks to this cabin ids, where T would be having the lowest rank and D would be having the hiest rank and so on..

dict_3 = {k:i for i,k in enumerate(survival_percantage)}
dict_3
```

```python
data["cabin_encoded"] = data["cabin_id"].map(dict_3)
```

```python
data
```

## Above in final column we can see, that we have encoded the cabin column successfully by giving it ranks based on the target variable

# 6. Mean Encoding - Nominal categories

It is exactly similar to above till the groupby function where we find the average of survived for every category. Here, we are going to replace that category with the mean directly.



```python
data.groupby(["cabin_id"])["Survived"].mean()
```

```python
dict_4 = data.groupby(["cabin_id"])["Survived"].mean().to_dict()
dict_4
```

```python
data["Cabin_mean_encoding"] = data["cabin_id"].map(dict_4)
```

```python
data # here, in last column, we have replaced cabin id with the mean of survived for that cabin specifically.
```

### Advantages : 
It captures information within the label, therefore rendering more predictive features.
Creates monotonic relationship between variable and a target.

### Disadvantage : 
prones to overfitting

# 7. Probability Ratio Encoding - Nominal Categories





## We are going to use the same titanic dataset and cabin column here to explain this enconding technique.
## Till now, we have replaced cabins with ranks, mean;now here, we are going to replace every category in cabin with probability ratio of survived to non-survived for that specific cabin.




```python
data
```

```python
survived_ratio = data.groupby(["cabin_id"])["Survived"].mean() # This represents the ratio of people survived to total people (probability of survived).
survived_ratio
```

```python
# Lets convert this into new dataframe

survived_df = pd.DataFrame(survived_ratio)
survived_df
```

```python
survived_df["Not_survived"] = 1 - survived_df["Survived"]
```

```python
survived_df # Here, we calculated the probability of not_survived
```

```python
# Now we are going to find the ratio of prob of survived to prob of not_survived.

survived_df["ratio"] = survived_df["Survived"]/survived_df["Not_survived"]
survived_df
```

```python
# Now we are going to use this ratio to replace the cabin values. 

# Lets convert ratio into dictionary first and then we would map it to cabin values

ratio_dictionary = survived_df["ratio"].to_dict()
ratio_dictionary
```

```python
# Lets map this dictionary to cabin data(categorical data)

data["Cabin_probability_ratio_encoding"] = data["cabin_id"].map(ratio_dictionary)
```
