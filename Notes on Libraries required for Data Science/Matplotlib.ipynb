plotting library for python programmming language. Its base library is numpy. Provides API to embed plots into application . Seaborn is also visualisation library and is imp in exploratory data analysis(stats analysis) , we use seaborn mostly. But here I have explained matplotlib.

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(10,20)
```

```python
# plotting using matplotlib

#plt scatter

plt.scatter(x,y,c="green")
plt.xlabel("Title for x axis")
plt.ylabel("Title for y axis")
plt.title("whole graphs title")
# plt.savefig("Text.png") #If you want to save image in system (it would get saved in jupyternotebooks' folder)
```

```python
y=x*x
```

```python
# plt plot function 

plt.plot(x,y)  # it gives in line format
```

```python
plt.plot(x,y,"g")  #adds color
```

```python
plt.plot(x,y,"g--") #adds green color and gives dashed line
```

```python
plt.plot(x,y,"g*--") # adds green color and then star and double dashed lines
```

```python
plt.plot(x,y,"go") #adds green color, then gives dotted (similar to scatter) 
```

```python
plt.plot(x,y,"g*", linestyle="dashed", linewidth=2, markersize=40) #adds green color, then gives dashed lines, then changes the width of that line, then markersize increses the size of markers (similar to scatter) 
```

```python
# creating subplots - it creates multiple plots in a given canvas

plt.subplot(2,2,1) # it means it would have 2 rows and 2 columns, and if we have mentioned 1 then plot would get plotted in the 1st place, lets see in diagram. 
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y,"g*")
plt.subplot(2,2,3)
plt.plot(x,y,"ro")
plt.subplot(2,2,4)
plt.plot(x,y,"pink",marker="*")
```

```python
x = np.arange(1,10)
y = 2*x+x**3
plt.plot(x,y,"g*--")
```

# Bar plot

```python
x = [1,2,5]
y = [4,5,6]

x2 = [4,6,8]
y2 = [5,10,15]

```

```python
plt.bar(x,y)
plt.bar(x2,y2,color="g")
plt.title("Bar chart")
plt.xlabel("title of x axis")
plt.ylabel("title of y axis")
```

# Histogram

```python
arr_for_hist = np.array([1,2,3,4,7,10,12,15,30,29,12,12,30])
plt.hist(arr_for_hist) #on x axis, the above numbers are present and on y axis the frequency of this numbers are present.
#By default bin value is 10, but if you want to change it you can
plt.title("Histogram plot")
```

```python
plt.hist(arr_for_hist, bins=20) # here bins represent in how many sections points from 1 to 30 would get divided into.(why 1 anfd 30, cause whatever points we have taken come in between 1 to 30). so here they would get divided into 20 different sections(check the array below .. 1., 2.45 ...there are 20 sections like this and it would then check how many points come in between 1 to 2.45 then again how many points would come in between 2.45 to 3.9)
```

# Box plot using Matplotlib

```python
# plot boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

data

plt.boxplot(data, vert=False, patch_artist= False) # here vert makes the boxplot vertical and patch_artist fills the color in boxplot. If you make them false boxplot would be plotted horizontal with no color
```

```python
plt.boxplot(data, vert=True, patch_artist= True)
```

# Piechart


```python
# plot pie chart

sizes= [100,215,340,600]
explode= [0.1,0,0,0]
labels= 'python', 'java', 'c++', 'javascript'
colors= ["green", "blue", "red", "pink"]

plt.pie(sizes, explode, labels, colors, autopct="%1.1f%%", shadow= True)

#sizes represent the numbers we are giving to plot in pie chart
#explode represents how far we want our slices to go away from centre
#labels represents labels for slices
#colors represnts the colrs we want for each slice
# autopct represents in which format you want to show your numbers in slices
#shadow would give shadow to piechart
```

```python
explode= [0.1,0,0,0.4]

plt.pie(sizes, explode, labels, colors, autopct="%1.1f%%")

plt.axis("equal")
```
