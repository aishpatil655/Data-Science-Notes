# It is used for visualisation. It provides alot of statistical tools which helps us undersatnd a lot of  things about data.

so in machine learning, we divide features into independent and dependent features (in supervised machine learning). Now if you want to see the relation between this features, we plot them. If we have 1 independent or 2 indepndent features, we can plot them but then plotting 3 or more features its diffidult to see the relation btween them by plotting. For that seasborn provides different functions which would help with plotting the relation between multiple features.

1. If we have single feature(dependent or independent), thats univariate analysis
2. If we have two features(any two), thats bivariate analysis
3. If we have multiple features, thats multivariate analysis.

This all things can be done with seaborn easily. seaborn library would help us with that by prviding mutiple plots like

1. distplot
2. joinplot- Will help with bivariate analysis. it would help to see relation between 2 features
3. pairplot- will help with multivariate analysis. It would help to see rlation between multiple fearures.( more than 2 features)

## Jointplot -Bivariate analysis

It allows to study the relationship between 2 numerical features. Central chart displays the correlation. It is usually a scatterplot, hexbin plot, 2d histogram or a 2d density plot.

```python
sns.jointplot(x="total_bill",y="tip", data=seaborn_numerical_data, kind="scatter") 

# it would show the distribution of individual features, as well as the correltaion between this features. 

#`kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'] - check the correlation plot inside a box, this options are for that plot.
```

```python
sns.jointplot(x="total_bill",y="tip", data=seaborn_numerical_data, kind="reg") 
```

# Pairplot- Multivariate analysis

used for multivariate analysis. Here, the relation between all numerical variables is shown. Uses scatterplot to show the correlation between 2 features. For relation of feature with itself, the histogram is shown which represents the distribution of that variable.

```python
sns.pairplot(seaborn_numerical_data) #here, to divide every individual feature based on sex(male and female), we can use hue. But understand, here we have used the data which had all categorical variables as well.otherwise it would not recognise sex feature.
```

```python
sns.pairplot(seaborn_data,hue="sex") # can use any categorical variable in hue.
```

# Dist plot

It is used to check the distribution of numerical columns/ features

```python
sns.distplot(seaborn_data["tip"]) 
```

```python
sns.distplot(seaborn_data["total_bill"]) 
```

```python
# If kde (kernel density estimation)  is present, this histrogram would have density of y axis. If kde is not present, it would show frequency/count onn y axis

sns.distplot(seaborn_data["tip"], kde=False, bins=20) # here, kde is false, so it would show frequency on y axis. Also, you can change number of bins. We have seen concept of bins above
```

# Boxplot

Boxplot shows the information in the form of 5 number summary (0 quantile, 25 quantile, 50 quantile, 75 quantile, 100 quantile)

```python
sns.boxplot(data=seaborn_data, x="sex", y="total_bill") # look at the data outside the boxplot, it represents outliers.

#compusory one variable must be numerical. both can be numerical, but would be messier(try).
```

```python
sns.boxplot(data=seaborn_data, x="day", y="total_bill", palette="rainbow")
```

```python
# boxplot for all numerical features

sns.boxplot(data=seaborn_data, orient="hor") # If you dont give any values for x or y, it would give boxplot for all numerical features.

#by default the orient is vertical.
```

```python
sns.boxplot(data=seaborn_data,x="day", y="total_bill", hue="sex") #here, you need to have 
```
